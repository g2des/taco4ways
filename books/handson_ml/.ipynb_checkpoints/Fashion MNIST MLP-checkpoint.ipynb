```python
import tensorflow as tf
```

```python
tf.__version__, tf.keras.__version__
```

```python
dataset = tf.keras.datasets.fashion_mnist
((Xtrain_full, ytrain_full), (Xtest, ytest)) = dataset.load_data()
Xtrain_full.shape, ytrain_full.shape, Xtest.shape, ytest.shape
```

```python
Xtrain_norm, Xvalid_norm = Xtrain_full[:50000]/255.0, Xtrain_full[50000:]/255.0
Xtrain, Xvalid = Xtrain_full[:50000], Xtrain_full[50000:]
ytrain, yvalid = ytrain_full[:50000], ytrain_full[50000:]
Xtest_norm = Xtest/255.0
(Xtrain.shape, Xtrain_norm.shape, ytrain.shape,
 Xtest.shape, ytest.shape,
 Xvalid.shape, Xvalid_norm.shape, yvalid.shape)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=[28,28]),
    tf.keras.layers.Dense(300, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')], name='mnist_fashion')
model_norm = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=[28,28]),
    tf.keras.layers.Dense(300, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')], name='mnist_fashion')
model.summary(), model_norm.summary()
```

```python
model.compile(loss=tf.keras.losses.sparse_categorical_crossentropy, optimizer=tf.keras.optimizers.SGD(learning_rate=1e-3),
             metrics=[tf.keras.metrics.sparse_categorical_accuracy])
model_norm.compile(loss=tf.keras.losses.sparse_categorical_crossentropy, optimizer=tf.keras.optimizers.SGD(learning_rate=1e-3),
             metrics=[tf.keras.metrics.sparse_categorical_accuracy])
```

```python
history = model.fit(Xtrain, ytrain, epochs=30, validation_data=(Xvalid, yvalid))
history_norm = model_norm.fit(Xtrain_norm, ytrain, epochs=30, validation_data=(Xvalid_norm, yvalid))
```

```python
model.evaluate(Xtest, ytest), model_norm.evaluate(Xtest_norm, ytest)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame(history.history)
df_norm = pd.DataFrame(history_norm.history)
df = df.join(df_norm, on=df.index, rsuffix='_norm')
df.plot(figsize=(16,10))
plt.gca().set_ylim(0,1)
plt.grid(True)
```

```python
Xnew = Xtest[:3]
yprob = model.predict(Xnew)
yclass = model.predict_classes(Xnew)
yprob.round(3), yclass, ytest[:3]
```
